```python
import geopandas as gpd
import matplotlib.pyplot as plt

national_park_checkpost_shpfile = "Bardia_Guard_Post84/"
national_park_shpfile = "Bardia_National_Park/"

# Read the GeoDataFrames from the shapefiles
gdf = gpd.read_file(national_park_checkpost_shpfile)
national_gdf = gpd.read_file(national_park_shpfile)

buffer = 0.02  # Assuming the buffer distance is 0.02 degrees (adjust as needed)

# Check the current CRS of the GeoDataFrame
print("Original CRS:", gdf.crs)

# Reproject to WGS 84 (EPSG:4326)
projected_crs = 'EPSG:4326'
gdf = gdf.to_crs(projected_crs)
national_gdf = national_gdf.to_crs(projected_crs)

# Create the buffer using the projected CRS
gdf['buffer'] = gdf.geometry.buffer(buffer)

# Plot the result
ax = national_gdf.plot(edgecolor='red', linewidth=2)
gdf['buffer'].plot(ax=ax, alpha=0.5, color='orange')
plt.title('Buffered Check Posts')
plt.show()
```
