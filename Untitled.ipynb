```python
import numpy as np
import pandas as pd
from scipy.stats import gamma

# Set a seed for reproducibility
np.random.seed(42)

# Generate random daily precipitation values for the period 1991-2020
start_date = "1991-01-01"
end_date = "2020-12-31"
date_range = pd.date_range(start=start_date, end=end_date, freq="D")
random_precipitation = np.random.gamma(shape=2, scale=5, size=len(date_range))

# Create a DataFrame with date and precipitation columns
precipitation_data = pd.DataFrame({"Date": date_range, "Precipitation": random_precipitation})

# Set the 'Date' column as the DataFrame index
precipitation_data.set_index("Date", inplace=True)

# Resample to weekly and calculate the sum
weekly_precipitation = precipitation_data.resample("W").sum()

# Extract the last 28 days of precipitation data
last_28_days = weekly_precipitation.iloc[-28:]

# Extract the weekly precipitation values for the last 28 days
last_28_days_values = last_28_days["Precipitation"].values

# Fit gamma distribution to the last 28 days of precipitation
shape, loc, scale = gamma.fit(last_28_days_values)

# Calculate SPI for the current week using the fitted gamma distribution
current_week_precipitation = weekly_precipitation.iloc[-1]["Precipitation"]
spi_value = (current_week_precipitation - loc) / scale

# Print the fitted gamma distribution parameters
print("Gamma Distribution Parameters - Shape:", shape, "Scale:", scale)

# Print the SPI value for the current week
print("SPI Value for the Current Week:", spi_value)
```
